```python
import numpy as np
from matplotlib import pyplot as plt

from sklearn.cluster import MiniBatchKMeans, KMeans
from sklearn.decomposition import FastICA

from operator import methodcaller
```

```python
overlapping_patches = get_patches(dm.image, step=10, size=20, mask=dm.mask)
```

```python
# random examples of patches
n_examples = 20
random_patches = [overlapping_patches[i] for i in np.random.randint(0,len(overlapping_patches), n_examples)]
show_gallery(random_patches, title='examples of input patches')
```

```python
X = np.column_stack([p.flat for p in overlapping_patches])

trfm = transform()
X_centered_normed = trfm.center_and_normalize(X)
X_centered_whitened = trfm.whiten(X_centered_normed)
```

```python
n_components = 40
# n_components = None

ica = FastICA(n_components=n_components, algorithm='parallel', whiten=True, 
        fun='logcosh', fun_args=None, max_iter=200, 
        tol=0.0001, w_init=None, random_state=None)
ica.fit(X_centered_normed.T);
```

```python
center_imgs = map(methodcaller('reshape', (20,20)), ica.components_)
show_gallery(center_imgs, 'ICA components using centered and whitened data')
```

```python
nonoverlapping_patches, (nrows_patch, ncols_patch), mask = get_patches(dm.image, step=20, size=20, mask=dm.mask)
```

```python
X = np.column_stack([p.flat for p in nonoverlapping_patches])
X1 = trfm.center_and_normalize(X)
S = ica.transform(X1.T)
features = np.zeros((nrows_patch, ncols_patch, 40))
features[mask] = S
```

```python
show_gallery([features[..., i] for i in range(40)], ncols=4, figsize=(20,25), cmap=plt.cm.seismic, 
             vmin=features.min(), vmax=features.max())
```

```python
n_components = 40

kmeans = MiniBatchKMeans(n_clusters=n_components, batch_size=10000)
kmeans.fit(X_centered_normed.T)

valid = np.where(np.bincount(kmeans.labels_) > 1)
nonempty_cluster_centers = kmeans.cluster_centers_[valid]

center_imgs = map(methodcaller('reshape', (20,20)), nonempty_cluster_centers)
show_gallery(center_imgs, figsize=(20,1), cmap=plt.cm.gray, 
             title='kmeans cluster centers using centered but not whitened data')
```

```python
labels = -1 * np.ones((nrows_patch, ncols_patch), dtype=np.int)
labels[mask] = kmeans.labels_

vis = label2rgb(labels)
plt.imshow(vis)
plt.axis('off')
plt.show()
```
