```python
%reload_ext autoreload
%autoreload 2

import sys
import os
import time

import numpy as np
import bloscpack as bp
import matplotlib.pyplot as plt
%matplotlib inline
import vtk
from vtk.util import numpy_support

sys.path.append(os.path.join(os.environ['REPO_DIR'], 'utilities'))
from utilities2015 import *
from annotation_utilities import *
from metadata import *
from vis3d_utilities import *
```

```python
brain_spec = dict(name='MD585', resolution='10.0um', vol_type='annotationAsScore')
stack = brain_spec['name']
```

Build annotation volumes from annotation files using: `annotation/construct_annotation_volume_from_annotation_files_v5.ipynb`

# Load annotation volumes and convert to meshes

```python
for name_s in all_known_structures_sided:
# for name_s in ['7N_L']:

    try:
        volume = DataManager.load_original_volume_v2(stack_spec=brain_spec, structure=name_s)
    except:
        sys.stderr.write("Error loading %s\n" % name_s)
        continue
        
    t = time.time()
    mesh = volume_to_polydata(volume=volume, num_simplify_iter=3, smooth=True)

    sys.stderr.write('time: %.2f\n' % (time.time() - t) ) # ~ 150s for outerContour; 93s for sp5

    mesh_fp = DataManager.get_mesh_filepath_v2(brain_spec=brain_spec, structure=name_s)
    create_parent_dir_if_not_exists(mesh_fp)
    save_mesh_stl(mesh, mesh_fp)
    upload_to_s3(mesh_fp)
```

```python
meshes_wrt_wholebrain_volResol = DataManager.load_meshes_v2(brain_spec=brain_spec, 
                                                            structures=all_known_structures_sided,
                                                            sided=True, 
                                                            levels=None)
```
